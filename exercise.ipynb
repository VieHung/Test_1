```python
def evaluate_f1_components(tp, fp, fn):
    if (isinstance(tp, int) and isinstance(fp, int) and isinstance(fn, int)):
        if (tp >= 0 and fp >= 0 and fn >= 0):
            precision = tp / (tp + fp)
            recall = tp / (tp + fn)
            f1_score = 2 * precision * recall / (precision + recall)
            print("precision is: " + str(precision))
            print("recall is: " + str(recall))
            print("f1-score is: " + str(f1_score))
        else:
            print("tp and fp and fn must be greater than or equal zero")

    elif (isinstance(tp, int) == 0):
        print("tp must be an integer")

    elif (isinstance(fp, int) == 0):
        print("fp must be an integer")

    elif (isinstance(fn, int) == 0):
        print("fn must be an integer")
```

```python
evaluate_f1_components(tp=2, fp=-2, fn=4)
```

```python
import math

def is_number(n):
    try:
        float(n)
    
    except ValueError:
        return False
    return True

def sigmoid(x):
    return 1 / (1 + pow(math.e, -x))

def relu(x):
    if (x >= 0): return x
    else: return 0

def elu(x):
    if (x >= 0): return x
    else: return 0.01*(pow(math.e,x) - 1)

def interactive_activation_function():
    x = input("Input x = ")
    name  = input("Input activation Function (sigmoid|relu|elu): ")
    if (is_number(x)):
        x = float(x)
        if (name in ['sigmoid', 'relu', 'elu']):
            if (name == 'sigmoid'):
                print(f"{name}: {str(sigmoid(x))}")
            elif (name == 'relu'):
                print(f"{name}: {str(relu(x))}")
            else:
                print(f"{name}: {str(elu(x))}")
        else:
            print(f"{name} is not supported")
    
    else:
        print("x is not a number")
```

```python
interactive_activation_function()
```

```python
import math
import random

def mae(predict, target, num_samples):
    mae = 0
    for i in range(num_samples):
        mae += abs(predict[i] - target[i])
    
    return mae / num_samples

def mse(predict, target, num_samples):
    mse = 0
    for i in range(num_samples):
        mse += pow(predict[i] - target[i],2)
    
    return mse / num_samples 

def rmse(predict, target, num_samples):
    return math.sqrt(mse(predict, target, num_samples))

def loss_functions():
    num_samples = input()
    
    if (num_samples.isnumeric()):
        target = []
        predict = []
        num_samples = int(num_samples)

        func_name = input()
        if (func_name in ['mae', 'mse', 'rmse']):
            for i in range(num_samples):
                target.append(random.uniform(0,10))
                predict.append(random.uniform(0,10))
            
            if (func_name == 'mae'):
                print(f"{func_name}: {str(mae(predict, target, num_samples))}")
            elif (func_name == 'mse'):
                print(f"{func_name}: {str(mse(predict, target, num_samples))}")
            else:
                print(f"{func_name}: {str(rmse(predict, target, num_samples))}")
        else:
            print(f"{func_name} is not supported")        
        
    else:
        print("number of samples must be an integer number")
```

```python
def factorial(n):
    if (n <= 0): return 1
    else: return factorial(n-1) * n

def approx_sin(x, n):
    sin_x = 0
    for i in range(n + 1):
        temp = pow(-1,i)*pow(x,2*i + 1)/factorial(2*i + 1)
        if (abs(temp) < 1e-15): return sin_x
        sin_x += temp
    return sin_x

def approx_cos(x, n):
    cos_x = 0
    for i in range(n + 1):
        temp = pow(-1,i)*pow(x,2*i)/factorial(2*i)
        if (abs(temp) < 1e-15): return cos_x
        cos_x += temp
    return cos_x

def approx_sinh(x, n):
    sinh_x = 0
    for i in range(n + 1):
        temp = pow(x, 2*i + 1)/factorial(2*i + 1)
        if (abs(temp) < 1e-15): return sinh_x
        sinh_x += temp
    return sinh_x

def approx_cosh(x, n):
    cosh_x = 0
    for i in range(n + 1):
        temp = pow(x, 2*i)/factorial(2*i)
        if (abs(temp) < 1e-15): return cosh_x
        cosh_x += temp
    return cosh_x
```

```python
# ex 2.2.2:

def count_character(word):
    """
    Đếm số lần xuất hiện của từng ký tự trong chuỗi đầu vào.

    Parameters:
        word (str): Chuỗi ký tự cần thống kê.

    Returns:
        dict: Từ điển với mỗi ký tự là một khóa, giá trị là số lần xuất hiện.
    """
    character_statistic = {}

    for character in word:
        if character in character_statistic:
            character_statistic[character] += 1
        else:
            character_statistic[character] = 1

    return character_statistic

# Kiểm tra hàm
assert count_character("Baby") == {'B': 1, 'b':1, 'a': 1, 'y': 1}

if __name__=="__main__":
    print(count_character('smiles'))
```

```python
# ex 2.2.3:

# Download file
# !gdown https://drive.google.com/uc?id=1IBScGdW2xlNsc9v5zSAya548kNgiOrko

def preprocess_text(sentence):
    """
    Tiền xử lý một câu bằng cách:
    - Chuyển tất cả các ký tự thành chữ thường
    - Loại bỏ dấu chấm (.) và dấu phẩy (,)
    - Tách câu thành danh sách các từ

    Parameters:
        sentence (str): Chuỗi câu đầu vào.

    Returns:
        list: Danh sách các từ sau khi xử lý.
    """
    sentence = sentence.lower()
    sentence = sentence.replace('.', '').replace(',', '')
    words = sentence.split()
    return words

def count_word(file_path):
    """
    Đếm số lần xuất hiện của từng từ trong văn bản đầu vào sau khi tiền xử lý.

    Quá trình tiền xử lý gồm:
    - Chuyển văn bản thành chữ thường
    - Loại bỏ dấu chấm (.) và dấu phẩy (,)
    - Tách văn bản thành danh sách từ

    Parameters:
        file_path (str): Đường dẫn đến file.

    Returns:
        dict: Từ điển đếm số lần xuất hiện của từng từ.
    """
    with open(file_path, 'r') as f:
        document = f.read()
    words = preprocess_text(document)

    counter = {}
    for word in words:
        if word in counter:
            counter[word] += 1
        else:
            counter[word] = 1

    return counter

if __name__=="__main__":
    file_path = './P1_data.txt'
    result = count_word(file_path)
    assert result.get('who', 0) == 3
    print(result.get('man', 0))
```
